# **Two Sum (Soma de Dois)**

Problema de Array

Link do LeetCode: [Problema Two Sum](https://leetcode.com/problems/two-sum/description/)

**Descrição:**

Dado um array de `nums` (números) inteiros e um `target` (alvo) inteiro, retornar índices dos dois números cuja soma seja igual ao target.

É possível considerar que cada input teria apenas uma solução possível, e que não é possível utilizar o mesmo elemento duas vezes.

É possível retornar a respostas em qualquer ordem.

````python
# Dado um array de nums (números) inteiros e um target (alvo) inteiro, retornar indices dos dois numeros cuja soma seja igual ao target, sem repetir elementos

nums = [5, 2, 39, 17, 30, 4, 9, 20]
target = 32

print('Array:', nums)
print('Target:', target)

def twoSum(nums, target):
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)): # Corrigido: o loop interno deve começar de i + 1
            if nums[i] + nums[j] == target:
                return [i, j]

resultado = twoSum(nums, target)

print('Indices dos dois números:', resultado)
````

## **Identificando a Complexidade do Código**

### **Complexidade de Tempo**

O código está utilizando dois loops for, um dentro do outro. Isto significa que, para cada número na lista, o código está percorrendo a lista inteira novamente procurando o par.

Isto pode se tornar um problema, já que, caso tivermos um array de 10 números, seriam feitas até 100 comparações.

Esta é a **Complexidade Quadrática ($O(n^2)$)**, funcional para listas pequenas, porém problemático quando estamos trabalhando com grandes volumes de dados.

### **Complexidade de Memória**

Como neste código estamos apenas usando duas variáveis (`i` e `j`) para controlar os índices, e não criando novas listas ou estruturas que guardam todos os dados novamente, o espaço usado na memória é fixo, não importando se o array tiver 10 ou 1000 itens.

Portanto, esta é a **Complexidade Constante ($O(1)$)**.

Assim, podemos dizer que o código é lento (Tempo), porém econômico (Memória).

## **2º Prompt**

Como posso tornar o seguinte código mais eficaz, sabendo que a Complexidade de Tempo é Quadrática, e a Complexidade de Memória é Constante? Mexa apenas no que for necessário e que diga respeito à essa questão.

`[Código]`

### **Resposta:**

Para tornar o código mais eficaz, vamos aplicar o que chamamos de **Hash Map** (que no Python é o `dict`).

A estratégia muda de: *"Para cada número, procure em todo o resto do array"* para: *"Para cada número, verifique se o que falta para chegar no alvo já foi visto antes"*.

Aqui está o código otimizado:

````python
# Versão Otimizada: Tempo O(n) e Memória O(n)

nums = list(map(int, input("Digite os números do array separados por espaço: ").split()))
target = int(input("Digite o target: "))

print('Array:', nums)
print('Target:', target)

def twoSum(nums, target):
    # Dicionário para guardar {valor: indice}
    mapa_indices = {}

    for i, num in enumerate(nums):
        complemento = target - num

        # Se o complemento já estiver no dicionário, achamos a dupla!
        if complemento in mapa_indices:
            return [mapa_indices[complemento], i]

        # Caso contrário, guarda o número atual e seu índice no dicionário
        mapa_indices[num] = i

resultado = twoSum(nums, target)

print('Índices dos dois números:', resultado)
````

#### **O que mudou e por que ficou melhor?**



1.   **Apenas um Loop** (`O(n)`): Agora percorremos a lista apenas uma vez. Em vez de um segundo `for` (que era lento), fazemos uma busca no dicionário `if complemento in mapa_indices`.
2.   **Busca Instantânea:** Em Python, verificar se uma chave existe em um dicionário leva tempo constante ($O(1)$). É por isso que o tempo total cai de quadrático para **linear**.
3.   **O "Trade-off" de Memória:** Como agora usamos o `mapa_indices` para memorizar os números que já passaram, a complexidade de memória subiu para $O(n)$. Sacrificamos um pouco de RAM para ganhar muita velocidade de processamento.

# **Reverse Linked List (Lista Encadeada Inversa)**

Problema de Lista

Link do LeetCode: [Problema do Reverse Linked List](https://leetcode.com/problems/reverse-linked-list/description/)

**Descrição:**

Dado o `head` de uma lista encadeada, inverter a lista e retornar *a lista invertida*.

````python
# Dado o `head` de uma lista encadeada, inverter a lista e retornar *a lista invertida*.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# criando a lista encadeada
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

# função para inverter
def reverseList(head):
  anterior = None
  atual = head

  while atual is not None:
    proximo = atual.next
    atual.next = anterior
    anterior = atual
    atual = proximo
    return anterior

def printLinkedList(head):
  while head is not None:
    print(head.val, end=" -> ")
    head = head.next
  print("None")


print("Lista encadeada original:")
printLinkedList(head)

head = reverseList(head)

print("\nLista encadeada invertida:")
printLinkedList(head)
````

### **Resposta:**
O erro no seu código é um detalhe sutil, mas muito comum em Python: a **indentação**.

**O Diagnóstico**

O seu comando `return anterior` está "dentro" do bloco do `while`.

Isso significa que, assim que o código entra no loop pela primeira vez (com o nó de valor 1), ele executa a inversão do primeiro ponteiro e encontra o `return`. Na mesma hora, a função para de funcionar e entrega o que tem em mãos. Por isso você só vê o `1 -> None`.

**A Solução**

Você precisa "tirar" o `return` de dentro do loop, para que ele só seja executado **depois** que o `while` terminar de percorrer toda a lista.

Veja como o bloco da função deve ficar:



```
def reverseList(head):

    atual = head

        atual = proximo       # Move o 'atual' um passo à frente
    
    # O return deve estar alinhado com o 'while', e não dentro dele!
    return anterior
```



````python
# Dado o `head` de uma lista encadeada, inverter a lista e retornar *a lista invertida*.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# criando a lista encadeada
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(5)

# função para inverter
def reverseList(head):
  anterior = None
  atual = head

  while atual is not None:
    proximo = atual.next
    atual.next = anterior
    anterior = atual
    atual = proximo

  return anterior


def printLinkedList(head):
  while head is not None:
    print(head.val, end=" -> ")
    head = head.next
  print("None")


print("Lista encadeada original:")
printLinkedList(head)

head = reverseList(head)

print("\nLista encadeada invertida:")
printLinkedList(head)
````

## **Identificando a Complexidade do Código**

### **Complexidade de Tempo**

Como para inverter a lista inteira precisamos passar por cada índice apenas uma vez, caso a lista tenha 5 elementos, acontecerão 5 trocas, e assim por diante. Assim, o tempo cresce na mesma proporção que o tamanho a lista, portanto, esta é a **Complexidade Linear ($O(n)$)**.

### **Complexidade de Memória**

Como não foi criada uma nova lista, e sim apenas uma reorganização da mesma lista, foram necessárias apenas as três referências utilizadas: `anterior`, `atual` e `proximo`. Como o tamanho da lista não faria nenhuma diferença, o gasto de memória é **fixo**, portanto, esta é a **Complexidade Constante ($O(1)$)**.

# **Simplify Path (Caminho Simplificado)**

Problema de Pilha

Link do LeetCode: [Problema do Simplify Path](https://leetcode.com/problems/simplify-path/)

**Descrição:**

Você recebeu um caminho *absoluto* para um sistema de arquivos no estilo Unix, que sempre começa com uma barra `'/'`. Sua tarefa é transformar esse caminho absoluto em seu **caminho canônico simplificado**.

As *regras* de um sistema de arquivos no estilo Unix são as seguintes:



*   Um único ponto `'.'` representa o diretório atual.
*   Um ponto duplo `'..'` representa o diretório anterior/diretório pai.
*   Múltiplas barras consecutivas, como `'//'` e `'///'`, são tratadas como uma única barra `'/'`.
*   Qualquer sequência de pontos que **não corresponda** às regras acima deve ser tratada como um **nome de diretório ou arquivo válido**. Por exemplo, `'...'` e `'....'` são nomes válidos de diretório ou arquivo.


O caminho canônico simplificado deve seguir estas *regras*:


*   O caminho deve começar com uma única barra `'/'`.
*   Os diretórios dentro do caminho devem ser separados por exatamente uma barra `'/'`.
*   O caminho não deve terminar com uma barra `'/'`, a menos que seja o diretório raiz.
*   O caminho não deve conter pontos simples ou duplos (`'.'` e `'..'`) usados ​​para denotar diretórios atuais ou diretórios pais.


Retorne o **caminho canônico simplificado**.

**Restrições:**



*   `1 <= path.length <= 3000`
*   `path` consiste em letras do alfabeto inglês, dígitos, ponto `'.'`, barra `'/'` ou `'_'`.
*   `path` é um caminho absoluto Unix válido.






````python
# Transformar o caminho de arquivos no sistema Unix para o simplificado
def simplificarCaminho(caminho):
    stack = []
    partes = caminho.split('/')

    for parte in partes:
          if parte == '' or parte == '.':
              continue
          if parte == '..':
              if stack:
                  stack.pop()
          else:
              stack.append(parte)

    return "/" + "/".join(stack)

print(simplificarCaminho("/home/"))
print(simplificarCaminho("/home//foo/"))
print(simplificarCaminho("/home/user/Documents/../Pictures"))
print(simplificarCaminho("/../"))
print(simplificarCaminho("/.../a/../b/c/../d/./"))
````

## **Identificando a Complexidade do Código**

### **Complexidade de Tempo**
Para simplificar o caminho, o código precisa executar três passos principais: separar a string original em partes (utilizando a barra `/` como guia), então percorrer cada uma dessas partes uma única vez para decidir se deve empilhar ou desempilhar um nome, e, então, juntar tudo novamente.

Como cada caractere e cada diretório são processados apenas uma vez, o tempo de execução cresce na mesma proporção que o tamanho do caminho fornecido. Portanto, esta é a **Complexidade Linear ($O(n)$)**.


### **Complexidade de Memória**
Ao contrário dos problemas do Two Sum e do Reverse Linked List, em que o gasto de memória era fixo, neste problema é necessária uma estrutura auxiliar (pilha) para armazenar os nomes dos diretórios.

Caso o caminho for muito longo e cheio de pastas novas como `/a/b/c/d...`, a pilha precisará guardar todos os nomes simultaneamente. Além disso, a lista gerada ao separar a string também ocupa espaço, proporcional à entrada.

Como o gasto de memória aumenta conforme o tamanho do caminho cresce, esta é a Complexidade **Linear ($O(n)$)**.

Assim, é possível dizer que o código é equilibrado, pois mantém uma execução rápida (Tempo), porém exige um pouco mais de espaço para organizar os dados (Memória).

# **Remove Nth Node From End of List (Remover n-ésimo nó de lista)**

Problema de Lista Encadeada

Link do LeetCode: [Problema do Remove Nth Node From End of List](https://leetcode.com/problems/remove-nth-node-from-end-of-list/description/)

**Descrição:**

Dado o `head`de uma lista encadeada, remover o `n`-ésimo nó do final da lista e retornar o `head`.

````python
# Dado o `head`de uma lista encadeada, remover o `n`-ésimo nó do final da lista e retornar o `head`
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeNthFromEnd(head, n):
    dummy = ListNode(0)
    dummy.next = head

    slow = dummy
    fast = dummy

    for _ in range(n+1):
        fast = fast.next

    while fast:
        slow = slow.next
        fast = fast.next

    slow.next = slow.next.next

    return dummy.next
````

````python
# Testando
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
n = 2

def printLinkedList(head):
    current = head
    while current is not None:
        print(current.val, end=" -> ")
        current = current.next
    print("None")

print("Lista Original:")
printLinkedList(head)

head = removeNthFromEnd(head, n)

print("\nLista Modificada:")
printLinkedList(head)
````

## **Identificando a Complexidade do Código**

### **Complexidade de Tempo**
Para encontrar o nó especifico que deve ser removido, sem ter que medir o tamanho total da lista primeiro, o código utiliza dois ponteiros: o `fast`e o `slow`.

O `fast`abre uma vantagem, enquanto o `slow`o acompanha, garantindo que a lista seja percorrida uma única vez (one-pass). Como o número de operações cresce de forma direta conforme o tamanho da lista aumenta, esta é a **Complexidade Linear ($O(n)$)**.

### **Complexidade de Memória**
Embora tenham sido criados um nó extra (o `dummy`) e os ponteiros de controle, essa quantidade de memória adicional é fixa. Não importa se a lista tem 5 ou 5.000 elementos, o código sempre usará apenas esse pequeno conjunto de referências auxiliares para ajustar os ponteiros existentes. Como o gasto de memória não depende do tamanho da entrada, esta é a **Complexidade Constante ($O(1)$)**.
